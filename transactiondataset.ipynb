```python
import pandas as pd
df=pd.read_csv("tekworks/transactions.csv")
df
```

```python
# Calculate total sales amount
total_sales = df["t_amt"].sum()
print("Total Sales Amount:", total_sales)
```

```python
#
df["t_date"] = pd.to_datetime(df["t_date"])
df["year_month"] = df["t_date"].dt.to_period("M")
monthly_sales = df.groupby("year_month")["t_amt"].sum()
highest_month = monthly_sales.idxmax()
highest_amount = monthly_sales.max()
print("Month with Highest Total Transaction Amount:", highest_month)
print("Highest Total Amount:", highest_amount)
```

```python
# Calculate average transaction amount per customer
avg_per_customer = df.groupby("cust_id")["t_amt"].mean()
print(avg_per_customer)
```

```python
trend=df.groupby("year_month")["t_amt"].sum()
print(trend)
```

```python
#highest single transaction amount recorded
df['t_amt'].max()
```

```python
service_revenue = df.groupby("services")["t_amt"].sum()

# Find service with highest revenue
top_service = service_revenue.idxmax()

print("Service Category with Highest Revenue:", top_service)
```

```python
#product generated highest revenue
product_avg = df.groupby("products_used")["t_amt"].mean()
print(product_avg.idxmax())
print(product_avg.max())
```

```python
#average transaction amount vary between services
average_transaction=df.groupby("services")["t_amt"].mean()
average_transaction
```

```python
# Count unique customers
df["cust_id"].nunique()
```

```python
#customers have spent the most overall 
customers=df.groupby("cust_id")["t_amt"].sum()
customers
top_customer = customers.sort_values(ascending=False)
top_customer
print("Top Customer:", top_customer.idxmax())
print("Total Spent:", top_customer.max())
```

```python
# Calculate average number of transactions per customer
transactions_per_customer = df.groupby("cust_id").size()
avg_transactions = transactions_per_customer.mean()
print("Average Number of Transactions per Customer:", avg_transactions)
```

```python
#identifying customners who made purchase in defferent categories
customers=df.groupby("cust_id")["products_used"].nunique()
customers
```

```python
# Identify customers with more than one transaction (repeat buyers)
transactions_per_customer = df.groupby("cust_id").size()
total_customers = transactions_per_customer.count()
repeat_customers = transactions_per_customer[transactions_per_customer > 1].count()
repeat_percentage = (repeat_customers / total_customers) * 100
print("Percentage of Repeat Buyers:", repeat_percentage)
```

```python
#identifying product category with the highest sales
highest_sales=df.groupby("products_used")["t_amt"].sum()
highest_sales
print(highest_sales.idxmax())
print(highest_sales.max())
```

```python
#identifying services which are most popular by transaction count
service_count = df["services"].value_counts()
service_count
print(service_count.idxmax())
print(service_count.max())
```

```python
services=df.groupby("services")["t_amt"].mean()
services
print(services.idxmax())
print(services.max())
```

```python
#Identify the state with the highest total sales
total_sales=df.groupby("state")["t_amt"].sum()
total_sales
print(total_sales.idxmax())
print(total_sales.max())
```

```python
## Identify the city with the highest transaction count
highest_number=df["city"].value_counts()
highest_number
print(highest_number.idxmax())
print(highest_number.max())
```

```python
# Calculate the average transaction amount for each state
average_transaction=df.groupby("state")["t_amt"].mean()
average_transaction
```

```python
## Count number of transactions for each service in every state
services = df.groupby(["state", "services"]).size().unstack(fill_value=0)
print(services)
print("\nMost Popular Service in Each State:")
print(services.idxmax(axis=1))
```

```python
outdoor_sales = df.groupby("state")["t_amt"].sum()
outdoor_sales
print(outdoor_sales.idxmax())
print(outdoor_sales.max())
```

```python
# Identify the state with highest Outdoor Recreation sales
print(df["products_used"].unique())
outdoor_df = df[df["products_used"].str.strip().str.lower() == "outdoor recreation".lower()]
outdoor_sales = outdoor_df.groupby("state")["t_amt"].sum()
if not outdoor_sales.empty:
    print(outdoor_sales.idxmax())
    print(outdoor_sales.max())
else:
    print("No Outdoor Recreation data found.")
```

```python
# Compare average spending between the two states
state_df = df[df["state"].isin(["California", "Texas"])]
avg_spending = state_df.groupby("state")["t_amt"].mean()
print("Average Spending Comparison (California vs Texas):")
print(avg_spending)

```

```python
#Identify the quarter with highest sales
df["t_date"] = pd.to_datetime(df["t_date"])
df["quarter"] = df["t_date"].dt.to_period("Q")
quarter_sales = df.groupby("quarter")["t_amt"].sum()
top_quarter = quarter_sales.idxmax()
top_sales = quarter_sales.max()
print("Quarter with Highest Sales:", top_quarter)
print("Highest Sales Amount:", top_sales)
```

```python
df["t_date"] = pd.to_datetime(df["t_date"])
# Extract month (Year-Month format)
df["year_month"] = df["t_date"].dt.to_period("M")
# Calculate total sales per month
monthly_sales = df.groupby("year_month")["t_amt"].sum()
print("Total Sales Month by Month:")
print(monthly_sales)
```

```python
#identifying total number of transactions per month
total_number_transactions=df.groupby("year_month")["t_amt"].size()
total_number_transactions
```

```python
#Identify the season with highest sports equipment sales
df["t_date"] = pd.to_datetime(df["t_date"])
sports_df = df[df["products_used"].str.contains("Sports", case=False, na=False)].copy()
sports_df["quarter"] = sports_df["t_date"].dt.to_period("Q")
sports_sales = sports_df.groupby("quarter")["t_amt"].sum()
print(sports_sales)
print("Highest Season:", sports_sales.idxmax())
print("Highest Sales:", sports_sales.max())
```

```python
# Count total number of such transactions
credit_transactions = df[df["t_details"] == "Credit"].shape[0]
print("Number of Credit Transactions:", credit_transactions)
```

```python
# Calculate total revenue generated from these transactions
credit_revenue = df[df["t_details"] == "Credit"]["t_amt"].sum()
credit_revenue
```

```python
# Compute average transaction amount for each payment type
avg_payment = df.groupby("t_details")["t_amt"].mean()
avg_payment
```

```python
# Identify top-performing regions based on revenue
state_sales = df.groupby("state")["t_amt"].sum().sort_values(ascending=False)
city_sales = df.groupby("city")["t_amt"].sum().sort_values(ascending=False)
print("Top States:")
print(state_sales.head())
print("\nTop Cities:")
print(city_sales.head())
```

```python
# Count total number of purchases for this category
fitness_count = df[df["products_used"] == "Exercise & Fitness"].shape[0]
print("Number of Exercise & Fitness Transactions:", fitness_count)
```

```python
# Identify categories generating high volume but lower per-transaction value
category_sales = df.groupby("products_used")["t_amt"].sum()
category_avg = df.groupby("products_used")["t_amt"].mean()
result = pd.DataFrame({
    "Total_Sales": category_sales,
    "Average_Transaction": category_avg
}).sort_values("Total_Sales", ascending=False)

print(result)
```
